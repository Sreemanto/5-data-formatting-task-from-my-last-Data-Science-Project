```python
def generate_excel(dfs,file_names):
    for df,file_name in zip(dfs,file_names):
        df.to_excel(file_name+'.xlsx',index=False)
```

```python
def get_data(Input_file,Output_files):
    Analysts_Availability = pd.DataFrame()
    
    Analysts = pd.read_json(Input_file).iloc[:,1:]
    
    for df in pd.read_json(Input_file).iloc[:,0]:
        Analysts_Availability = pd.concat([Analysts_Availability,pd.DataFrame(df)])

    return generate_excel([Analysts_Availability, Analysts],Output_files)
```

```python
Input_file = 'analyst_in_tool.json'
Output_files = ['Analysts_Availability','Analysts']

get_data(Input_file,Output_files)
```
